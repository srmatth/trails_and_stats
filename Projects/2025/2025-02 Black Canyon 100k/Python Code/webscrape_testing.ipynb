```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
driver.get('https://live.aravaiparunning.com/#/black_canyon-2025')
```

```python
first_names = driver.find_elements(By.CLASS_NAME, "first-name")
f_names = [name.text for name in first_names]
last_names = driver.find_elements(By.CLASS_NAME, "last-name")
l_names = [name.text for name in last_names]
gps = driver.find_elements(By.CLASS_NAME, "gp")
gender_pos = [gp.text for gp in gps]
bibs = driver.find_elements(By.CLASS_NAME, "bib")
bib_no = [bib.text for bib in bibs]
```

```python
first_names = driver.find_elements(By.CLASS_NAME, "first-name")
# urls = []

# data = []

for i in range(len(urls), len(first_names)):

    first_names = driver.find_elements(By.CLASS_NAME, "first-name")
    # print(len(first_names))
    runner = first_names[i]
    # Click the runner's name
    runner.click()
    
    # Wait for the new page to load (adjust time as necessary)
    driver.implicitly_wait(10)
    
    # Get the current URL (which is the runner's individual page URL)
    runner_url = driver.current_url
    urls.append(runner_url)
    print(f"Runner URL: {runner_url}")

    li_elements = driver.find_elements(By.CSS_SELECTOR, "li.has-crossing")

    # Loop through each <li> and extract the required data
    for li in li_elements:
        # Extract the split name
        split_name = li.find_element(By.CSS_SELECTOR, "span.split-name").text
    
        # Extract the time of day
        time_of_day = li.find_element(By.CSS_SELECTOR, "span.tod").text
    
        # Extract the distance
        distance = li.find_element(By.CSS_SELECTOR, "span.dist").text
    
        # Extract the duration (hours, minutes, seconds, ms)
        hours = li.find_elements(By.CSS_SELECTOR, "span.hours")
        minutes = li.find_element(By.CSS_SELECTOR, "span.minutes").text
        seconds = li.find_element(By.CSS_SELECTOR, "span.seconds").text
        milliseconds = li.find_element(By.CSS_SELECTOR, "span.ms").text
    
        # Check if hours element exists, and extract it if so
        if hours:
            hours = hours[0].text
        else:
            hours = '00'

        # Create a dictionary for the current row of data
        row = {
            "First Name": f_names[i],
            "Last Name": l_names[i],
            "Bib": bib_no[i],
            "Gender Pos": gender_pos[i],
            "Split Name": split_name,
            "Time of Day": time_of_day,
            "Distance": distance,
            "Hours": hours,
            "Minutes": minutes,
            "Seconds": seconds,
            "Milliseconds": milliseconds
        }
    
        # Append the row to the data list
        data.append(row)
    
    # Go back to the original page to continue with the next runner
    driver.back()
    driver.implicitly_wait(10)

# Close the browser
driver.quit()
```
